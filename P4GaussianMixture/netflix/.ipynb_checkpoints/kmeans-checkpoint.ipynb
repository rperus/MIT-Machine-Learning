```python
def estep(X: np.ndarray, mixture: GaussianMixture) -> Tuple[np.ndarray, float]:
    n, d = X.shape
    mu, var, pi = mixture 
    K = mu.shape[0]
    
    post = (np.exp(-np.linalg.norm(X[:,None] - mu, ord=2, axis=2)**2/(2*var))/(2*np.pi*var)**(d/2)*pi)/(np.sum((np.exp(-np.linalg.norm(X[:,None] - mu, ord=2, axis=2)**2/(2*var))/(2*np.pi*var)**(d/2)*pi), axis=1).reshape(-1,1))    
    
    log = np.sum(np.log((np.sum((np.exp(-np.linalg.norm(X[:,None] - mu, ord=2, axis=2)**2/(2*var))/(2*np.pi*var)**(d/2)*pi), axis=1).reshape(-1,1))), axis=0).item()    
    
    return post, log
```

```python
def mstep(X: np.ndarray, post: np.ndarray) -> GaussianMixture:

    n, d = X.shape
    K = post.shape[1]
    
    mu = (post.T @ X)/(np.sum(post, axis=0).reshape(-1,1))   
    var = np.sum(post*(np.linalg.norm(X[:, None] - mu, ord=2, axis=2)**2), axis=0)/(np.sum(post, axis=0)*d)   
    p = np.sum(post, axis=0)/n 
    
    return GaussianMixture(mu, var, p)
```

```python
def run(X: np.ndarray, mixture: GaussianMixture,
        post: np.ndarray) -> Tuple[GaussianMixture, np.ndarray, float]:

    prev_cost = None
    cost = None  
    

    while (prev_cost is None or cost - prev_cost >= (1e-6*np.abs(cost))):
        prev_cost = cost
#         post = estep(X, mixture)
#         mixture, cost = mstep(X, post)
        post, cost = estep(X, mixture)
        mixture = mstep(X, post)
            
    return mixture, post, cost
```

```python
run(X, GaussianMixture,post)
```

```python
def bic(X: np.ndarray, mixture: GaussianMixture,
        log_likelihood: float) -> float:
    """Computes the Bayesian Information Criterion for a
    mixture of gaussians

    Args:
        X: (n, d) array holding the data
        mixture: a mixture of spherical gaussian
        log_likelihood: the log-likelihood of the data

    Returns:
        float: the BIC for this mixture
    """
    n = X.shape[0]    
    p = 0
    
    for i in range(len(mixture)):
        if i == 0:
            p = mixture[i].shape[0] * mixture[i].shape[1] +p
        else:
            p = mixture[i].shape[0]   +p
    p = p - 1
    
    return  log_likelihood - (p*np.log(n))/2.0
    
    
```

```python
bic(X, GaussianMixture, p)
```
